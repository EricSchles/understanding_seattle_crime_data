So far we can state the following conclusions regarding the dataset:
1. CAD CDW ID is not a unique identifier - this means that some IDs are repeated which could lead to collisions.  Typically in police work this isn't _that_ big a deal because the case IDs are no longer referenced, however this will become probablematic overtime for any longer term analysis.  Therefore it is my recommendation that a new ID be created with an eye towards longer term analysis.
2. CAD Event is not a unique identifier either - the same recommendation as for CAD CDW ID
3. General Offense is not unique either - the same recommendation as for CAD CDW ID as well.

Clearance Code is unrelated to CAD CDW ID - it appears to be a semantic identifier however it doesn't appear to match up with the Event Clearance Description because Event Clearance Description contains 124 unique entries while Clearance Code has 13655 unique entries.  

Clearance Code does not appear to match up with Event Clearance SubGroup either.
Event Clearance SubGroup and Event Clearance Description _do_ appear to match up.  The same number of entries have a nan value.  That cannot be coincidence.  
Event Clearance SubGroup and Event Clearance Group have the same number of nan's they are clearly match up directly as well.

Understanding the Dataset:
It appears as though the word ALACAD refers to a false alarm.  The way I was able to discern this is from the above reference where CAD is defined and the first 3 letters of ALA being the same as the first letters of alarm.  Since ALACAD is mentioned in reference with false alarms, as shown below, I think it's a safe bet these are alarms in the CAD system.  This could be an initial point of analysis.  Predicting how many False alarms occur each year.  According to https://www.seattle.gov/police/community-policing/false-alarm-program this is a big issue for the department.  Costing around 1 million dollars a year.

Let's see how well an ARIMA model does at predicting the total number of False alarms per year.  We'll do this for each category of False Alarms:

1. PANIC ALARMS (FALSE)
2. VEHICLE ALACAD (FALSE)
3. BURGLARY ALARMS (FALSE)
4. VEHICLE ALARMS (FALSE)
5. BURGLARY ALACAD (FALSE)
6. PANIC ALACAD (FALSE)


```python
#Cleaning and Preparing the Data
#Group by month: http://stackoverflow.com/questions/33350381/average-of-daily-count-of-records-per-month-in-a-pandas-dataframe
tmp_df = df[pd.notnull(df["Event Clearance Date"])]
tmp_df["Event Clearance Date"] = pd.to_datetime(tmp_df["Event Clearance Date"], format="%m/%d/%Y %I:%M:%S %p")
panic_alarms_df = tmp_df[tmp_df["Event Clearance SubGroup"] == "PANIC ALARMS (FALSE)"]
daily = panic_alarms_df.set_index('Event Clearance Date').groupby(pd.TimeGrouper(freq="D"))["Event Clearance SubGroup"].count()
monthly_panic_alarms = daily.groupby(pd.TimeGrouper(freq="M")).sum()
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(monthly_panic_alarms.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(monthly_panic_alarms, lags=40, ax=ax2)
```

Based on the above chart, we'll make use of the partial autocorrelation function, because it appears to degrade more quickly.  And therefore we'll use an AR process with a sliding window of 2.

```python
monthly_panic_alarms = monthly.astype(float)
panic_model = sm.tsa.ARIMA(monthly_panic_alarms, (2,0,0)).fit(disp=0)
panic_model.fittedvalues
ax = monthly_panic_alarms.plot(figsize=(12,8))
ax = panic_model.fittedvalues.plot(ax=ax, style='r--', label='Fitted Values')
```

As we can see here visually our fitted values do a decent job approximating our data, which is a good thing!  Next we'll look at some summary statistics to quantify how good of a job our model is doing.

```python
print("DW test",sm.stats.durbin_watson(panic_model.resid.values))
print(stats.normaltest(panic_model.resid))
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
fig = qqplot(panic_model.resid, line='q',ax=ax,fit=True)
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(panic_model.resid.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(panic_model.resid, lags=40, ax=ax2)
```
